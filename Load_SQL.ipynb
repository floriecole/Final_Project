```python
#Dependencies
import pandas as pd
import numpy as np
import pymysql
from sqlalchemy import create_engine
from sqlalchemy import Column, Integer, String, Float
pymysql.install_as_MySQLdb() 
```

```python
#Read the csv file into a pandas DataFrame
books_df = pd.read_csv('Resources/BX-Books.csv',encoding="ISO-8859-1")
books_df.head()
```

```python
books_df.shape
```

```python
#Remove unwanted characters using Regex
books_df['publisher']=books_df['publisher'].replace('&amp;', '', regex=True)
books_df
```

```python
# Drop all rows with missing information
books_df1 = books_df.dropna()
books_df1
```

```python
usrratings_df = pd.read_csv('Resources/BX-Book-Ratings.csv',encoding="ISO-8859-1")
usrratings_df.head()
```

```python
# Drop all rows with missing information
usrratings_df1 = usrratings_df.dropna()
usrratings_df1
```

```python
reads_df = pd.read_csv('Resources/good_reads_final.csv')
reads_df.head()
```

```python
# Drop all rows with missing information
reads_df1 = reads_df.dropna()
reads_df1
```

```python
#Remove unwanted characters using Regex
reads_df['author_name']=reads_df['author_name'].replace('\n', '', regex=True)
reads_df['birthplace']=reads_df['birthplace'].replace('\n', '', regex=True)
reads_df['book_title']=reads_df['book_title'].replace('\n', '', regex=True)
reads_df
```

```python
users_df = pd.read_csv('Resources/BX-Users.csv',encoding="ISO-8859-1")
users_df.head()
```

```python
# Drop all rows with missing information
users_df = users_df.dropna()
users_df
```

```python
#Merge tables
merge1_df = pd.merge(users_df, usrratings_df, left_on='userID', right_on='userID', how='inner')
merge1_df.shape
```

```python
merge1_df.head()
```

```python
merge2_df = pd.merge(usrratings_df, books_df, left_on='ISBN', right_on='ISBN', how='inner')
merge2_df.shape
```

```python
merge2_df.head()
```

```python
merge3_df = pd.merge(books_df, reads_df, left_on='bookAuthor', right_on='author_name', how='inner')
merge3_df.shape
```

```python
merge3_df.head(20)
```

```python
#Drop duplicates
books_all_df = merge3_df.drop_duplicates(subset=["ISBN"]) 
books_all_df.head()
```

```python
books_all_df.shape
```

```python
# Drop all rows with missing information
books_all_df1 = books_all_df.dropna()
books_all_df1.head()
```
